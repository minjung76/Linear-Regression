```python
import pandas as pd
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)

pd_DataSet=pd.read_csv("./Prediction_Sample.csv")
```

```python
DF=pd.DataFrame(pd_DataSet)
```

```python
New_DF=DF.dropna(how='any')
```

```python
print(New_DF)
```

```python
import sklearn.model_selection as sms
```

```python
x=New_DF[['X1','X2']]
```

```python
y=New_DF['Y1']
```

```python
x_train,x_test,y_train,y_test=sms.train_test_split(x,y,train_size=0.8,test_size=0.2)
```

```python
import sklearn.linear_model as slm

lr_Model=slm.LinearRegression()
lr_Model.fit(x_train,y_train)
```

```python
predict_X=lr_Model.predict(x)
```

```python
import sklearn.metrics as slt

MEA_score=slt.mean_absolute_error(predict_X,y)
print('MAE:',MEA_score)  #평균 #극단적인 값에 영향을 받는다.

```

```python
MSE_score=slt.mean_squared_error(predict_X,y)
print('MSE:',MSE_score) #분산 #극단적인 값에 영향을 받는다.
```

```python
RMSE_Score=MSE_score ** 0.5 

print(RMSE_Score) #표준편차 0에 가까울수록 예측 정확도가 높다 #극단적인 값에 영향을 받는다.
```

```python
import numpy as np

np_RMSE_score=np.sqrt(MSE_score)


print(np_RMSE_score) #표준편차
```

```python
#R2_score 상대적 분산 기반 검증 정확한 퍼센테이지를 알려준다. 
#극단적인 값에 영향을 받지 않는다. 

R2_score=slt.r2_score(predict_X,y)
print('R2_score',R2_score)

#0.9969767859086441 약 99프로가 정확하다. 
#예측을 하였으면 정확도를 확인해야 한다. 
```
